```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
e = 0.7
a = 4 # in terms of Rs
w = 250*(np.pi/180) # rad
i = 80*(np.pi/180) # rad
```

```python
def coord_wrt_alpha(alpha, a, e, w, i):
    angle = alpha + w
    r = a * (1 - e**2) / (1 + e*np.cos(alpha))
    x = r * np.cos(angle)
    y = r * np.sin(angle) * np.cos(i)
    return x, y
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

for x in range(360):
    alpha = x*(np.pi/180)
    x, y = coord_wrt_alpha(alpha, a, e, w, i)
    if x**2 + y**2 < 1.0:
#         print(x,y)
        ax.scatter(x,y,c='g',s=1)
    else:
        ax.scatter(x,y,c='k',s=1)

ax.set_xlim(-8,8)
ax.set_ylim(-8,8)

c = patch.Circle((0,0), 1.0, zorder=0, color='r', fill=None)
ax.add_patch(c)
```
